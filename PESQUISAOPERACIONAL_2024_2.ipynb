Problema P1:

Uma f√°brica produz dois tipos de produtos: A e B. Para produzir cada unidade do produto A, s√£o necess√°rias 2 horas de m√£o de obra e 1 kg de mat√©ria-prima. Problema P2:

Uma pequena empresa produz dois tipos de produtos: X e Y. Para produzir cada unidade do produto X, s√£o necess√°rias 3 horas de m√£o de obra e 2 kg de mat√©ria-prima. Para o produto Y, s√£o necess√°rias 2 horas de m√£o de obra e 3 kg de mat√©ria-prima. A empresa disp√µe de 60 horas de m√£o de obra e 50 kg de mat√©ria-prima por semana. Problema P3:

Uma pequena marcenaria produz dois tipos de m√≥veis: cadeiras e mesas. Para produzir cada cadeira, s√£o necess√°rias 2 horas de trabalho e 1 metro quadrado de madeira. Para cada mesa, s√£o necess√°rias 3 horas de trabalho e 2 metros quadrados de madeira. A marcenaria disp√µe de 48 horas de trabalho e 30 metros quadrados de madeira por semana. **Conceitos**

Opera√ß√£o - (Estrat√©gica, T√°tica e Operacional)

Restri√ß√µes

Solu√ß√£o √≥tima (Max, Min)

Constantes

Vari√°veis

````python
import numpy as np
import matplotlib.pyplot as plt

# Definir as restri√ß√µes como fun√ß√µes
def restricao1(x, y):
  return 2*x + 1*y <= 40

def restricao2(x, y):
  return 1*x + 2*y <= 30

# Criar um meshgrid para avaliar as restri√ß√µes em todos os pontos
x = np.linspace(0, 30, 100)
y = np.linspace(0, 20, 100)
X, Y = np.meshgrid(x, y)

# Avaliar as restri√ß√µes no meshgrid
Z1 = restricao1(X, Y)
Z2 = restricao2(X, Y)

# Plotar as regi√µes vi√°veis
plt.contourf(X, Y, Z1, colors=['white', 'gray'], alpha=0.5)
plt.contourf(X, Y, Z2, colors=['white', 'gray'], alpha=0.5)

# Plotar as linhas das restri√ß√µes
plt.contour(X, Y, Z1, levels=[0], colors='black')
plt.contour(X, Y, Z2, levels=[0], colors='black')

# Plotar a fun√ß√£o objetivo (opcional, para visualizar a dire√ß√£o do aumento)
Z = 10*X + 15*Y
plt.contour(X, Y, Z, levels=[0, 500, 1000, 1500], colors='red')

# Configurar o gr√°fico
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regi√£o Vi√°vel e Fun√ß√£o Objetivo')
plt.grid(True)
plt.show()
````
