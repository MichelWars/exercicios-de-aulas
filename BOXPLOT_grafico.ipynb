```python
import pandas as pd
import matplotlib.pyplot as plt

dados = pd.DataFrame({'Local': ['Bairro 1', 'Bairro 2'],
                      'Habitantes': [1500, 1800],
                      'Homens': [900, 400]})

# Criar boxplots para todas as colunas
plt.figure(figsize=(10, 5))  # Definir tamanho da figura

# Boxplot para a coluna 'Habitantes'
plt.subplot(1, 2, 1)  # Subplot 1
dados.boxplot(column='Habitantes')
plt.title('Boxplot de Habitantes')
plt.ylabel('Habitantes')

# Boxplot para a coluna 'Homens'
plt.subplot(1, 2, 2)  # Subplot 2
dados.boxplot(column='Homens')
plt.title('Boxplot de Homens')
plt.ylabel('Homens')

plt.tight_layout()  # Ajustar o layout

plt.show()
```
